```python
import os
import pandas as pd
import numpy as np
```

```python
source_directory="./data/raw/"
data_set=[]
seasons=[]

for x in os.listdir(source_directory):
    data_set.append(x)
    seasons.append(x.strip('.csv'))

data_set.sort()
seasons.sort()

# Create empty dict to keep track of which clubs appear and how many times
club_seasons={}
```

```python
for x in data_set:
    season_index=data_set.index(x)
    season=seasons[season_index]
    
    # Read in csv as pandas dataframe
    df=pd.read_csv(source_directory+x)
    
    # Isolate club names from dataframe
    club_list=df['HomeTeam'].unique()
    clubs=[]
    
    for i in club_list:
        if i in clubs:
            continue
        else:
            clubs.append(i)

    clubs.sort()
    
    for i in clubs:
        if i in club_seasons:
            club_seasons[i]+=1
        else:
            club_seasons[i]=1
        
    # Track which clubs have played so the values can be added to the 2d array
    match_count= np.zeros((len(clubs),len(clubs)),int)
    
    # Create empty 2d array to hold yellows and reds for each ref
    yc_data = np.zeros((len(clubs),len(clubs)), int)
    
    # Fill the 2d array with the card count for each ref and club
    for i in range(0,len(df['AY'])):
        home_club=df['HomeTeam'][i]
        away_club=df['AwayTeam'][i]
        yc_data[clubs.index(home_club),clubs.index(away_club)]+=df['HY'][i]
        yc_data[clubs.index(away_club),clubs.index(home_club)]+=df['AY'][i]
    
    temp_df=pd.DataFrame(data=yc_data, index=clubs, columns=clubs)
    #temp_df.to_csv(season+'.csv')
    
    if data_set.index(x)==0:
        master_df=temp_df
    else:
        master_df=pd.concat([master_df,temp_df]).groupby(level=0).sum()
```
